```python
from rdkit import Chem
from rdkit.Chem import AllChem, DataStructs
import pandas as pd

def get_top_similar(df, query_smile, top_n=20):
    # Convert query SMILES to mol and fingerprint
    query_mol = Chem.MolFromSmiles(query_smile)
    query_fp = AllChem.GetMorganFingerprintAsBitVect(query_mol, radius=2)

    # Compute similarity for each SMILES in df
    similarities = []
    for smile in df['Smiles']:
        mol = Chem.MolFromSmiles(smile)
        if mol is not None:
            fp = AllChem.GetMorganFingerprintAsBitVect(mol, radius=2)
            sim = DataStructs.TanimotoSimilarity(query_fp, fp)
            similarities.append(sim)
        else:
            similarities.append(0.0)  # Assign 0 if invalid SMILES

    df['similarity'] = similarities
    top_df = df.sort_values(by='similarity', ascending=False).head(top_n)
    return top_df
```

```python
from rdkit.Chem import Draw

def visualize_top_similar(df, query_smile, top_n=20):
    top_df = get_top_similar(df, query_smile, top_n)

    # Generate RDKit Mol objects
    mols = [Chem.MolFromSmiles(smi) for smi in top_df['Smiles']]
    legends = [f"Sim: {sim:.2f}, pChEMBL: {pchembl:.2f}" for sim, pchembl in zip(top_df['similarity'], top_df['pChEMBL Value'])]


    # Draw molecules in a grid
    img = Draw.MolsToGridImage(mols, molsPerRow=3, subImgSize=(400, 200), legends=legends)
    return img
```
